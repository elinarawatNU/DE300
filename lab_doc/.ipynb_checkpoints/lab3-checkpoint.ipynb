# Lab3: EDA


Exploratory Data Analysis (EDA) is a critical first step in analyzing the data from an experiment or data project. EDA allows analysts to:
- Understand the data's underlying patterns.
- Spot anomalies.
- Test a hypothesis or check assumptions.
- Build an intuition for the data.


## Data Loading

```python

import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris_data = load_iris()
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)

# Display the first few rows of the dataframe
iris_df.head()
```

## Summary Statistics

```python

# Summary statistics for continuous variables
display(iris_df.describe())

# Frequency counts for categorical variables
display(iris_df['species'].value_counts())
```

## Data Visualization

```python

import matplotlib.pyplot as plt

# Histograms for continuous variables
iris_df.hist(figsize=(12, 8), bins=15)
plt.suptitle('Histograms of Continuous Variables')
plt.show()

# Bar chart for the categorical variable
iris_df['species'].value_counts().plot(kind='bar')
plt.title('Frequency of Each Species')
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.show()
```

## Identifying Relationships Between Variables

```python

# Scatter plot matrix for continuous variables
sns.pairplot(iris_df, hue='species', diag_kind='kde')
plt.suptitle('Scatter Plot Matrix for Continuous Variables')
plt.show()

# Box plots for continuous variables by categorical variable
plt.figure(figsize=(12, 8))
for i, col in enumerate(iris_data.feature_names, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=col, data=iris_df)
    plt.title(f'Box Plot of {col} by Species')
plt.tight_layout()
plt.show()
```
